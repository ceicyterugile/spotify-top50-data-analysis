```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

outlier_indices = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    col_outlier_indices = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
    if not col_outlier_indices.empty:
        outlier_indices[col] = col_outlier_indices.tolist()

if outlier_indices:
    print("Potential Outliers (Indices):")
    for col, indices in outlier_indices.items():
        print(f"- {col}: {indices}")
        display(df.loc[indices, [col]])
else:
    print("No potential outliers found using the IQR method.")
```

```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_df.corr()
strongly_positive_corr = correlation_matrix[correlation_matrix > 0.7].stack().reset_index()
strongly_positive_corr.columns = ['Feature 1', 'Feature 2', 'Correlation Coefficient']
strongly_positive_corr = strongly_positive_corr[strongly_positive_corr['Feature 1'] != strongly_positive_corr['Feature 2']]
strongly_positive_corr = strongly_positive_corr[
    strongly_positive_corr.apply(lambda row: (row['Feature 2'], row['Feature 1']) not in zip(strongly_positive_corr['Feature 1'], strongly_positive_corr['Feature 2']), axis=1)
]

if not strongly_positive_corr.empty:
    print("Features with strong positive correlation (> 0.7):")
    display(strongly_positive_corr)
else:
    print("No strong positive correlations found among numerical features (> 0.7).")
```

No pairs of numerical features have a correlation coefficient greater than 0.7, meaning that within the Top 50 tracks, there isn't a strong linear tendency for any two numerical audio features to increase together significantly.

That means there is no single dominant combination of features. Hit songs in this list achieve their popularity through a variety of audio profiles, rather than relying on one specific combination of highly correlated features.

Popularity here seems to be achieved through a more diverse set of audio characteristics.

```python
strongly_negative_corr = correlation_matrix[correlation_matrix < -0.7].stack().reset_index()
strongly_negative_corr.columns = ['Feature 1', 'Feature 2', 'Correlation Coefficient']
strongly_negative_corr = strongly_negative_corr[strongly_negative_corr['Feature 1'] != strongly_negative_corr['Feature 2']]
strongly_negative_corr = strongly_negative_corr[
    strongly_negative_corr.apply(lambda row: (row['Feature 2'], row['Feature 1']) not in zip(strongly_negative_corr['Feature 1'], strongly_negative_corr['Feature 2']), axis=1)
]


if not strongly_negative_corr.empty:
    print("Features with strong negative correlation (< -0.7):")
    display(strongly_negative_corr)
else:
    print("No strong negative correlations found among numerical features (< -0.7).")
```

A very low correlation coefficient (close to 0, either positive or negative) between two features means that there is little to no linear relationship between them.

Features with very low correlation are largely independent of each other in terms of their linear movement. For example, the low correlation between 'Danceability' and 'Liveness' (around -0.0066) suggests that a song's danceability level doesn't have a linear relationship with how "live" it sounds (audience noise, etc.) in this top 50 list.

The presence of many uncorrelated features further supports the idea that there isn't a simple formula for a hit song based on just a couple of linearly related audio properties. Success in the top 50 seems to be influenced by a combination of factors that aren't necessarily linked together in a straightforward linear fashion.

Similar to the lack of strong positive correlations, the presence of many low correlations suggests that analyzing the distribution and typical values of each feature individually, or in combination with categorical data like genre, might be more insightful than focusing solely on linear relationships between numerical features.
In essence, the features with very low correlation are telling that these particular audio characteristics tend to vary quite independently in the top 50 tracks. This highlights the diverse nature of popular music in this dataset and suggests that a hit song's appeal comes from a mix of audio attributes that aren't simply scaling up or down together.

```python
import numpy as np

low_correlation_threshold = 0.1
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
low_corr_pairs = upper.stack().reset_index()
low_corr_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation Coefficient']
not_correlated_features = low_corr_pairs[abs(low_corr_pairs['Correlation Coefficient']) < low_correlation_threshold]

if not not_correlated_features.empty:
    print(f"Features with very low correlation (absolute correlation < {low_correlation_threshold}):")
    display(not_correlated_features)
else:
    print(f"No features found with correlation coefficient magnitude less than {low_correlation_threshold}.")
```

**Key insights about what makes a hit song in this Top 50 Spotify tracks dataset**

Artist and Album Influence: Artists like Dua Lipa, Billie Eilish, and Travis Scott have multiple tracks in the Top 50, suggesting that artist popularity plays a significant role. Similarly, certain albums having multiple popular tracks indicates that a successful album can contribute to individual song popularity.
Danceability is Key: A high proportion of the Top 50 tracks have a danceability score above 0.7, indicating that being danceable is a common characteristic among these popular songs. The most danceable track, "WAP," also holds a high rank, reinforcing this. However, the presence of tracks with lower danceability suggests that it's not the only factor.
Loudness Variation: While there are very loud tracks in the Top 50, there's also a significant range in loudness, including some quite quiet tracks. This suggests that extreme loudness isn't a strict requirement for a hit, and different genres or styles might have different typical loudness levels.
Duration Diversity: The Top 50 includes both relatively short and long tracks, indicating that there isn't a single ideal track length for a hit song.
Genre Dominance and Diversity: Pop and Hip-Hop/Rap are the most represented genres, highlighting their current popularity. However, the presence of 16 different genres in a list of 50 tracks shows that the Top 50 is not dominated by just a few genres and that popular music has considerable genre diversity.
Lack of Strong Linear Feature Relationships: The absence of strong positive or negative correlations between most numerical audio features suggests that there isn't a simple linear formula for a hit song based on these features alone. Hit songs in this dataset seem to achieve popularity through a variety of combinations of audio characteristics, rather than a few features always increasing or decreasing together. This means that analyzing the individual characteristics and their distributions, as well as their relationship with genre, is important.
In summary, while artist and album popularity, as well as high danceability, appear to be common themes, the Top 50 is diverse in terms of other audio features and genres. This suggests that "what makes a hit song" is a complex interplay of factors beyond simple linear relationships between the numerical audio characteristics provided in this dataset. Further analysis of how these features vary within and across genres would likely provide deeper insights.
